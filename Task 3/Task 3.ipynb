# Import libraries and load datasets

```python
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
from catboost import CatBoostRegressor, Pool
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
train_df = pd.read_csv('internship_train.csv')
test_df = pd.read_csv('internship_hidden_test.csv')
train_df.head()
```

# Plot features against the target

It can be noticed that probably only feature 6 is relevant for the given regression task. It requires to be further investigated.

```python
for column in train_df:
    plt.plot(train_df[column], train_df['target'], 'bo', markersize=3)
    plt.title("Feature {} against the target".format(column))
    plt.show()
```

# Train Catboost model

It can be noticed that the catboost model used only feature number 6 which confirms that it is most like the only informative feature, somehow correlated with the target.

```python
X_train, X_val, y_train, y_val = train_test_split(train_df.drop("target", axis="columns"), train_df["target"], test_size=0.2, random_state=0)

model = CatBoostRegressor(iterations=1000,
                          learning_rate=0.1,
                          depth=2,
                          eval_metric="RMSE")

eval_dataset = Pool(X_val, y_val)
model.fit(X_train, y_train, eval_set=eval_dataset, plot=True)


df_feature_importance = pd.DataFrame(model.get_feature_importance(prettified=True))
plt.figure(figsize=(12, 6))
feature_plot = sns.barplot(x="Importances", y="Feature Id", data=df_feature_importance, palette="cool");
plt.title('features importance')
plt.show()
```

# Train linear regression

Investigation of the relation between feature number 6 and target has indicated that it is quadratic relation. In order to model it, it is sufficient to include feature 6 squared. Such a procedure significantly simplifies the model and therefore increases it's interpretability. Obtained RMSE is the same as this obtained with Catboost, so dropping remaining features didn't contribute to decrease in the performance. 

```python
X_train, X_test, y_train, y_test = train_test_split(train_df.drop("target", axis="columns"), train_df["target"], test_size=0.2, random_state=0)

X_train = X_train[["6"]]
X_train['6^2'] = X_train['6'] ** 2

X_test = X_test[["6"]]
X_test['6^2'] = X_test['6'] ** 2

reg = LinearRegression().fit(X_train, y_train)
print("Model coefficients:", reg.coef_, reg.intercept_)

x1 = np.linspace(-10, 10, 1000, endpoint=True)

X = pd.DataFrame(columns=['6', '6^2'])
X["6"] = x1
X["6^2"] = x1 ** 2

plt.plot(X_train["6"], y_train, 'bo', label='Observation')
plt.plot(x1, reg.predict(X), 'r', label='Linear regression')
plt.title("Observations and trained linear model")
plt.xlabel("feature 6")
plt.ylabel("target")
plt.show()

y_pred = reg.predict(X_test)
residuals = y_test - y_pred
print("RMSE:", mean_squared_error(y_test, y_pred, squared=False))

plt.scatter(y_pred[:100], residuals[:100], )
plt.title("Residuals")
plt.xlabel("target")
plt.ylabel("residual error")
plt.show()
```

# Testing normality of residuals

P-value of Shapiro test in really small which indicates that most likely residuals do not have normal distribution. It is unfortunate since this assumption allows for more detailed probabilistic investigation of the model, like calculation of confidence intervals for coefficients of the linear model or for predictions. 

```python
from scipy import stats
shapiro_test = stats.shapiro(residuals[:5000])
shapiro_test
```

# Further investigation of residuals

Looking at the distribution of the residuals one might suspect that the noise is uniformly distributed. 
Final form of the model would look like the following:

f(x) = w1 * x ^ 2 + w2 * x + intercept + epsilon, where epsilon is a uniformly distributed noise

```python
print("Mean:", sum(residuals) / len(residuals))
print("Standard deviation:", np.std(residuals))
```

```python
plt.hist(residuals)
plt.title("Histogram of residuals")
plt.xlabel("residual error")
plt.ylabel("number of residuals")
plt.show() 
```

# Prediction on test dataset

```python
X_test = test_df[["6"]]
X_test['6^2'] = X_test['6'] ** 2
y_pred = reg.predict(X_test)
test_df["target"] = y_pred
test_df.to_csv('predictions.csv')
```
